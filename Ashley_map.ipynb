## Ashley Map

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy
```

```python
lons=[30.506165,30.812998,37.132532,37.165836,35.718022,35.816544,35.882228,34.1791667,35.948134,18.472829,30.4,36.652788]
lats=[29.594063,29.574761,12.508958,12.598916,3.853731,2.647792,2.346536,-0.402222,2.38412,25.797547,29.5,4.366125]
```

```python
plt.figure(figsize=(20,10))
ax=plt.axes(projection=ccrs.Robinson())
ax.set_global()
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)
ax.stock_img()
ax.set_extent((15,45,32,-5),ccrs.PlateCarree())

for n in range(0,len(lons)):
    plt.plot(lons[n],lats[n],marker='o',transform=ccrs.PlateCarree(),label='({},{})'.format(lons[n],lats[n]))

plt.legend(bbox_to_anchor=(1,0), loc="lower left")
```
