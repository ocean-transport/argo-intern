Compute spectra from this data

```python
import xrft 
```

```python
# we only consider the structure below the mixed layer
CT_data = ds_660.CT.sel(ctd_pressure=slice(200,950))
SA_data = ds_660.SA.sel(ctd_pressure=slice(200, 950))
```

```python
fig, axes = plt.subplots(figsize=(20,15))

#temperature,left
plt.subplot(2,2,1)
CT_data.plot(y='ctd_pressure')
plt.gca().invert_yaxis()

#salinity, right
plt.subplot(2,2,2)
SA_data.plot(y='ctd_pressure')
plt.gca().invert_yaxis()
```

```python
ds_660.CT.isel(dives=80).plot()
```

```python
fig, axes = plt.subplots(figsize=(20,15))

#temperature, left
plt.subplot(2,2,1)
(CT_data - CT_data.mean('dives')).plot(y='ctd_pressure')
plt.gca().invert_yaxis()

#salinity, right
plt.subplot(2,2,2,)
(SA_data - SA_data.mean('dives')).plot(y='ctd_pressure')
plt.gca().invert_yaxis()
```

```python
#temperature
CT_spec = xrft.power_spectrum(CT_data.dropna(dim='dives') - CT_data.mean('dives'), dim='ctd_pressure', window=True)

#salinity
SA_spec = xrft.power_spectrum(SA_data.dropna(dim='dives') - SA_data.mean('dives'),
                                     dim='ctd_pressure', window=True)
```

```python
fig, axes = plt.subplots(figsize=(20,15))

#temperature, left
plt.subplot(2,2,1)
CT_spec.mean('dives').plot(label='Spectra over full survey: TEMPERATURE', linewidth=2)
#CT_spec.sel(dives=slice(0,250)).mean('dives').plot(label='High EKE', linewidth=2)
#CT_spec.sel(dives=slice(250,500)).mean('dives').plot(label='Low EKE', linewidth=2)

kz = CT_spec.freq_ctd_pressure

plt.plot(kz, 3e-6*kz**(-2), '--', label='-2')
plt.plot(kz, 5e-8*kz**(-3), '--', label='-3')

plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$\lambda$ [1/m]')
plt.ylim([1e-8, 3])


#salinity, right
plt.subplot(2,2,2)
SA_spec.mean('dives').plot(label='Spectra over full survey: SALINITY', linewidth=2)
#SA_spec.sel(dives=slice(0,250)).mean('dives').plot(label='High EKE', linewidth=2)
#SA_spec.sel(dives=slice(250,500)).mean('dives').plot(label='Low EKE', linewidth=2)

kz = SA_spec.freq_ctd_pressure

plt.plot(kz, 3e-6*kz**(-2), '--', label='-2')
plt.plot(kz, 5e-8*kz**(-3), '--', label='-3')

plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$\lambda$ [1/m]')
plt.ylim([1e-8, 3])
```

```python
CT_spec.mean('dives').plot(label='Spectra over full survey: TEMPERATURE', figsize=(6,4))
kz = CT_spec.freq_ctd_pressure

plt.plot(kz, 3e-6*kz**(-2), '--', label='-2')
plt.plot(kz, 5e-8*kz**(-3), '--', label='-3')

plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$\lambda$ [1/m]')
plt.ylim([1e-8, 3])

#Where is the salinity label?
SA_spec.mean('dives').plot(label='Spectra over full survey: SALINITY')
kz = SA_spec.freq_ctd_pressure
```
