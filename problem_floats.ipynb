```python
from argopy import DataFetcher as ArgoDataFetcher
import argopy
import xarray as xr
```

```python
erddap_loader  = ArgoDataFetcher(src='erddap',parallel=True,progress=True)
ifremer_loader = ArgoDataFetcher(src='gdac',ftp='https://data-argo.ifremer.fr/',parallel=True,progress=True)
usgodae_loader = ArgoDataFetcher(src='gdac',ftp='ftp://usgodae.org/pub/outgoing/argo',parallel=True,progress=True)
balwada_loader = ArgoDataFetcher(src='gdac',ftp="/swot/SUM05/dbalwada/Argo_sync",parallel=True,progress=True)
```

```python
box_gyre=[-40,-30,20,30,0,2000]
box_medi=[-20,-10,35,45,0,2000]
box_equa = [-10,0,-10,0,0,2000] 
```

```python
ds = balwada_loader.region(box_medi)
print('loading points complete')
ds = ds.to_xarray()
print('to xarray complete')
ds = ds.argo.teos10(['CT','SA','SIG0'])
ds = ds.argo.point2profile()
print('point to profile complete')
```
