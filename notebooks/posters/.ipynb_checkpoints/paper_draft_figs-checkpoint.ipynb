```python
fig = plt.figure(figsize=(10, 4))
ax = plt.axes(projection=ccrs.PlateCarree())
lon_edges = np.arange(-180, 181, 3)
lat_edges = np.arange(-90, 91, 3)
heatmap, _, _ = np.histogram2d(ds.LATITUDE.values, ds.LONGITUDE.values, bins=[lat_edges, lon_edges])
lon_centers = (lon_edges[:-1] + lon_edges[1:]) / 2
lat_centers = (lat_edges[:-1] + lat_edges[1:]) / 2
X, Y = np.meshgrid(lon_centers, lat_centers)
start = 0
end   = 1000
norm  = colors.Normalize(vmin=start, vmax=end)
cmap  = cmo.gray_r
heatmap_plot = ax.pcolormesh(X, Y, heatmap, cmap=cmap, norm=norm, transform=ccrs.PlateCarree())
cbar = plt.colorbar(heatmap_plot, ax=ax, orientation='vertical')#, shrink=0.5)
cbar.set_label('Number of Profiles')
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.set_global()
ax.set_title(r'All Argo Profiles, $3^\circ$ x $3^\circ$ bins')
```

```python
# Define bin edges
lon_edges = np.arange(-180, 181, 3)
lat_edges = np.arange(-90, 91, 3)
lon_centers = (lon_edges[:-1] + lon_edges[1:]) / 2
lat_centers = (lat_edges[:-1] + lat_edges[1:]) / 2
X, Y = np.meshgrid(lon_centers, lat_centers)

# Depths to analyze
depths = [0, 100, 1000, 2000]
depth_indices = [np.argmin(np.abs(ds.PRES_INTERPOLATED.values - d)) for d in depths]

# Create figure
fig, axs = plt.subplots(2, 2, figsize=(20, 8), subplot_kw={'projection': ccrs.PlateCarree()})
axs = axs.flatten()

for i, (depth, idx) in enumerate(zip(depths, depth_indices)):
    # Mask valid profiles
    valid = (~np.isnan(ds.CT[:, idx])) & (~np.isnan(ds.SA[:, idx]))
    lat_valid = ds.LATITUDE.values[valid]
    lon_valid = ds.LONGITUDE.values[valid]

    # Bin into 2D histogram
    heatmap, _, _ = np.histogram2d(lat_valid, lon_valid, bins=[lat_edges, lon_edges])

    # Plot
    ax = axs[i]
    norm = colors.Normalize(vmin=0, vmax=1000)
    pcm = ax.pcolormesh(X, Y, heatmap, cmap=cmo.thermal, norm=norm, transform=ccrs.PlateCarree())
    ax.set_title(f'Valid CT & SA at {depth}m', fontsize=14)
    ax.coastlines()
    ax.gridlines(draw_labels=True)
    ax.set_global()

    # Colorbar
    cbar = plt.colorbar(pcm, ax=ax, orientation='vertical', shrink=0.7)
    cbar.set_label('Valid Profiles')

plt.tight_layout()
plt.suptitle('Argo Profile Coverage by Depth', fontsize=16, y=1.02)
plt.show()
```
