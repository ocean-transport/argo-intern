```python
ds = xr.open_dataset('/swot/SUM05/amf2288/sync-boxes/globe_binned_3_z.nc', chunks={'N_PROF':10000})
```

```python
ds = ds.rename({'z_c':'z', 'lat_c':'lat', 'lon_c':'lon'})
```

```python
from xgcm.autogenerate import generate_grid_ds
from xgcm import Grid
```

```python
ds_gridded = generate_grid_ds(ds, {'X':'lon', 'Y':'lat'})
ds_gridded = ds_gridded.drop_vars(['lat_l', 'lat_r', 'lon_l', 'lon_r'])
```

```python
grid = Grid(ds_gridded, coords={'X':{'center':'lon', 'left':'lon_left'},
                                'Y':{'center':'lat', 'left':'lat_left'},
                                'Z':{'center':'z'}},
                                periodic=['X'])
```

```python
dlong = grid.diff(ds_gridded.lon, 'X')
dlonc = grid.diff(ds_gridded.lon_left, 'X')

dlatg = grid.diff(ds_gridded.lat, 'Y', boundary='fill', fill_value=np.nan)
dlatc = grid.diff(ds_gridded.lat_left, 'Y', boundary='fill', fill_value=np.nan)
```

```python
dlonc = grid.diff(ds_gridded.lon_left, 'X')
print(dlatc)
```

```python
def dll_dist(dlon, dlat, lon, lat):
        """Converts lat/lon differentials into distances in meters

        PARAMETERS
        ----------
        dlon : xarray.DataArray longitude differentials
        dlat : xarray.DataArray latitude differentials
        lon  : xarray.DataArray longitude values
        lat  : xarray.DataArray latitude values

        RETURNS
        -------
        dx  : xarray.DataArray distance inferred from dlon
        dy  : xarray.DataArray distance inferred from dlat
        """

        distance_1deg_equator = 111000
        dx = dlon * np.cos(np.deg2rad(lat)) * distance_1deg_equator
        dy = ((lon * 0) + 1) * dlat * distance_1deg_equator
        return dx, dy
```

```python
ds_gridded.coords['dxg'], ds_gridded.coords['dyg'] = dll_dist(dlong, dlatg, ds_gridded.lon, ds_gridded.lat)
ds_gridded.coords['dxc'], ds_gridded.coords['dyc'] = dll_dist(dlonc, dlatc, ds_gridded.lon, ds_gridded.lat)
```

```python
dT_dx = grid.diff(ds_gridded.CT, 'X') / ds_gridded.dxg
dT_dy = grid.diff(ds_gridded.SA, 'Y') / ds_gridded.dyg
```

```python
fig=plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.set_global()
ds_gridded.dyc.plot(x='lon',y='lat',ax=ax,transform=ccrs.PlateCarree(), vmax=333000, vmin=150000)
ax.set_title('dyc')
```

```python
fig=plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.set_global()
ds_gridded.dxc.plot(x='lon',y='lat',ax=ax,transform=ccrs.PlateCarree(), vmax=333000, vmin=150000)
ax.set_title('dxc')
```

```python
depth=10

fig=plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.set_global()
dT_dx.sel(z=depth, method='nearest').plot(x='lon',y='lat',ax=ax,transform=ccrs.PlateCarree(), robust=True)
ax.set_title('dT_dx')
```

```python
depth=10

fig=plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.set_global()
dT_dy.sel(z=depth, method='nearest').plot(x='lon',y='lat',ax=ax,transform=ccrs.PlateCarree(), robust=True)
ax.set_title('dT_dy')
```

```python
dT_dx = grid.interp(dT_dx, ['X'], to='center')
dT_dy = grid.interp(dT_dy, ['Y'], to='center')
```

```python
temp_flux = np.sqrt(dT_dx**2 + dT_dy**2)
```

```python
depth=100

fig=plt.figure(figsize=(15,5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.gridlines(draw_labels=True)
ax.set_global()
temp_flux.sel(z=depth, method='nearest').plot(x='lon',y='lat',ax=ax,transform=ccrs.PlateCarree(), cmap=cmo.deep_r, norm=colors.LogNorm(vmin=1e-7, vmax=5e-6))
ax.set_title('$\sqrt{dT/dx^2 + dT/dy^2}$')
```
