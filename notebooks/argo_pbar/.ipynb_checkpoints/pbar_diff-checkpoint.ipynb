```python
ds_pbar_grid = xr.open_dataset('/swot/SUM05/amf2288/ds_pbar_grid_var.nc')
```

```python
grid = Grid(ds_pbar_grid, coords={
        'X': {'center': 'LON', 'left': 'LON_left'},
        'Y': {'center': 'LAT', 'left': 'LAT_left'},
        'Z': {'center': 'PRESSURE_mean'}}, 
        periodic=['X'], boundary={'Y':'fill'})
```

```python
K_rho_p = xr.open_dataarray('/swot/SUM05/amf2288/K_rho_filt_p.nc').drop_vars('Z').rename({'PRESSURE':'PRESSURE_mean'})
K_rho_p = K_rho_p.bfill(dim='PRESSURE_mean').ffill(dim='PRESSURE_mean')
```

```python
def get_grad(ds_gridded, grid, var):
    # --- horizontal grid spacing ---
    dlong = grid.diff(ds_gridded.LON, 'X')
    dlatg = grid.diff(ds_gridded.LAT, 'Y', boundary='fill', fill_value=np.nan)

    distance_1deg_equator = 111000
    dx = dlong * np.cos(np.deg2rad(ds_gridded.LAT)) * distance_1deg_equator
    dy = ((ds_gridded.LON * 0) + 1) * dlatg * distance_1deg_equator

    # --- horizontal gradients ---
    dC_dx = grid.diff(ds_gridded[var], 'X') / dx
    dC_dy = grid.diff(ds_gridded[var], 'Y', boundary='fill', fill_value=np.nan) / dy

    dC_dx = grid.interp(dC_dx, 'X', to='center')
    dC_dy = grid.interp(dC_dy, 'Y', to='center')

    # --- vertical gradient (pressure coordinate) ---
    p = ds_gridded.PRESSURE_mean
    dp = p.diff('PRESSURE_mean')
    dp = dp.broadcast_like(ds_gridded[var])

    dC_dp = ds_gridded[var].diff('PRESSURE_mean') / dp

    # restore original coordinate length without duplicates
    dC_dp = dC_dp.reindex(PRESSURE_mean=ds_gridded.PRESSURE_mean)

    return dC_dx, dC_dy, dC_dp
```

## Calculate Gradients

```python
ds_pbar_grid['dCT_m_dx'], ds_pbar_grid['dCT_m_dy'], ds_pbar_grid['dCT_m_dp'] = get_grad(ds_pbar_grid, grid, 'CT_sort_m')
ds_pbar_grid['dSA_m_dx'], ds_pbar_grid['dSA_m_dy'], ds_pbar_grid['dSA_m_dp'] = get_grad(ds_pbar_grid, grid, 'SA_sort_m')
ds_pbar_grid['dSPICE_m_dx'], ds_pbar_grid['dSPICE_m_dy'], ds_pbar_grid['dSPICE_m_dp'] = get_grad(ds_pbar_grid, grid, 'SPICE_sort_m')
```

```python
ds_pbar_grid['dCT_e_dx'], ds_pbar_grid['dCT_e_dy'], ds_pbar_grid['dCT_e_dp'] = get_grad(ds_pbar_grid, grid, 'CT_sort_e')
ds_pbar_grid['dSA_e_dx'], ds_pbar_grid['dSA_e_dy'], ds_pbar_grid['dSA_e_dp'] = get_grad(ds_pbar_grid, grid, 'SA_sort_e')
ds_pbar_grid['dSPICE_e_dx'], ds_pbar_grid['dSPICE_e_dy'], ds_pbar_grid['dSPICE_e_dp'] = get_grad(ds_pbar_grid, grid, 'SPICE_sort_e')
```

```python
ds_pbar_grid['dCT_dx'], ds_pbar_grid['dCT_dy'], ds_pbar_grid['dCT_dp'] = get_grad(ds_pbar_grid, grid, 'CT_sort')
ds_pbar_grid['dSA_dx'], ds_pbar_grid['dSA_dy'], ds_pbar_grid['dSA_dp'] = get_grad(ds_pbar_grid, grid, 'SA_sort')
ds_pbar_grid['dSPICE_dx'], ds_pbar_grid['dSPICE_dy'], ds_pbar_grid['dSPICE_dp'] = get_grad(ds_pbar_grid, grid, 'SPICE_sort')
```

## Calculate Diffusivity

```python
ds_pbar_grid['K_e_CT'] = K_rho_p * ds_pbar_grid.CT_e_var/(ds_pbar_grid.dCT_m_dx**2 + ds_pbar_grid.dCT_m_dy**2)
ds_pbar_grid['K_e_SA'] = K_rho_p * ds_pbar_grid.SA_e_var/(ds_pbar_grid.dSA_m_dx**2 + ds_pbar_grid.dSA_m_dy**2)
ds_pbar_grid['K_e_SPICE'] = K_rho_p * ds_pbar_grid.SPICE_e_var/(ds_pbar_grid.dSPICE_m_dx**2 + ds_pbar_grid.dSPICE_m_dy**2)
```
