### Results 1

```python
depth = 250
shrink, pad = 0.9, 0.1

# --- Configuration for each subplot ---
fields = [
    ("CT_sort", 1, cmo.thermal, (0,20), "Conservative Temperature (°C)", "Conservative Temperature"),
    ("SA_sort", 1, cmo.haline, "robust", "Absolute Salinity (g/kg)", "Absolute Salinity"),
    ("dCT_m_dx", 1000, cmo.balance, "robust", "dCT_m/dx (°C/km)", "Zonal Grad. of CT_m"),
    ("dSA_m_dx", 1000, cmo.balance, "robust", "dSA_m/dx (°C/km)", "Zonal Grad. of SA_m"),
    ("dCT_m_dy", 1000, cmo.balance, "robust", "dCT_m/dy (°C/km)", "Meridional Grad. of CT_m"),
    ("dSA_m_dy", 1000, cmo.balance, "robust", "dSA_m/dy (°C/km)", "Meridional Grad. of SA_m"),
    ("dCT_m_dp", 1, cmo.balance, "robust", "dCT_m/dz (°C/m)", "Vertical Grad. of CT_m"),
    ("dSA_m_dp", 1, cmo.balance, "robust", "dSA_m/dz (°C/m)", "Vertical Grad. of SA_m"),
]

# --- Helper function to plot each panel ---
def plot_panel(ax, varname, scale, cmap, vlims, cbar_label, title):
    da = ds_pbar_grid[varname].sel(PRESSURE_mean=depth, method='nearest')*scale

    da.plot(x='LON', y='LAT', cmap=cmap, transform=ccrs.PlateCarree(), vmin=vlims[0] if isinstance(vlims, tuple) else None, vmax=vlims[1] if isinstance(vlims, tuple) else None,
        robust=True if vlims == "robust" else False, ax=ax, cbar_kwargs={'shrink': shrink, 'pad': pad, 'label': cbar_label})

    ax.coastlines()
    ax.gridlines(draw_labels=True, alpha=0.75)
    
    # Title inside the panel, in gray
    ax.set_title(title)

# --- Create figure and loop ---
fig, axes = plt.subplots(4, 2, figsize=(15, 15), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=-60)})
axes = axes.ravel()

for i, ax in enumerate(axes):
    var, scale, cmap, vlims, cbar_label, title = fields[i]
    plot_panel(ax, var, scale, cmap, vlims, cbar_label, title)

plt.subplots_adjust(wspace=0.2, hspace=0.1)
plt.savefig(f"quals_paper/supplemental/1_{ds_pbar_grid.sel(PRESSURE_mean=depth, method='nearest').PRESSURE_mean.data}.png")
print(f"depth: {ds_pbar_grid.sel(PRESSURE_mean=depth, method='nearest').PRESSURE_mean.data} m")
```

### Results 2

```python
lon = -170
shrink, pad = 1, 0.02

# --- Configuration for each subplot ---
fields = [
    ("CT_sort", 1, cmo.thermal, (1,20), "Conservative Temperature (°C)", "Conservative Temperature", "white"),
    ("SA_sort", 1, cmo.haline, "robust", "Absolute Salinity (g/kg)", "Absolute Salinity", "white"),
    ("dCT_m_dx", 1000, cmo.balance, "robust", "dCT_m/dx (°C/km)", "Zonal Grad. of CT_m", "k"),
    ("dSA_m_dx", 1000, cmo.balance, "robust", "dSA_m/dx (°C/km)", "Zonal Grad. of SA_m", "k"),
    ("dCT_m_dy", 1000, cmo.balance, "robust", "dCT_m/dy (°C/km)", "Meridional Grad. of CT_m", "k"),
    ("dSA_m_dy", 1000, cmo.balance, "robust", "dSA_m/dy (°C/km)", "Meridional Grad. of SA_m", "k"),
    ("dCT_m_dp", 1, cmo.balance, "robust", "dCT_m/dz (°C/m)", "Vertical Grad. of CT_m", "k"),
    ("dSA_m_dp", 1, cmo.balance, "robust", "dSA_m/dz (°C/m)", "Vertical Grad. of SA_m", "k"),
]

# --- Helper function to plot each panel ---
def plot_panel(ax, varname, scale, cmap, vlims, cbar_label, title, color, is_bottom_row):
    da = ds_pbar_grid[varname].sel(LON=lon, method='nearest')*scale

    da.plot(x='LAT', y='PRESSURE_mean', cmap=cmap, vmin=vlims[0] if isinstance(vlims, tuple) else None, vmax=vlims[1] if isinstance(vlims, tuple) else None,
        robust=True if vlims == "robust" else False, ax=ax, cbar_kwargs={'shrink': shrink, 'pad': pad, 'label': cbar_label})

    ax.set_xlim(-72, 58)
    ax.set_ylim(2000, 0)
    ax.set_ylabel("Pressure Mean (dbar)")

    # Handle x‑labels and tick labels
    if is_bottom_row:
        ax.set_xlabel("Latitude (°N)")
        ax.set_xticklabels(ax.get_xticklabels())
    else:
        ax.set_xlabel("")
        ax.set_xticklabels([])

    # Title inside the panel, in gray
    ax.text(0.5, 0.1, title, transform=ax.transAxes, ha='center', va='top', fontweight='bold', color=color, fontsize=plt.rcParams['axes.titlesize'])
    ax.set_title('')

# --- Create figure and loop ---
fig, axes = plt.subplots(4, 2, figsize=(15, 15))
axes = axes.ravel()

for i, ax in enumerate(axes):
    var, scale, cmap, vlims, cbar_label, title, color = fields[i]
    is_bottom_row = i >= 6   # panels 6 and 7 (0‑indexed) are bottom row
    plot_panel(ax, var, scale, cmap, vlims, cbar_label, title, color, is_bottom_row)

plt.subplots_adjust(wspace=0.35, hspace=0.1)
plt.savefig(f"quals_paper/supplemental/2_{ds_pbar_grid.sel(LON=lon, method='nearest').LON.data}.png")
print(f"lon: {ds_pbar_grid.sel(LON=lon, method='nearest').LON.data} m")
```

```python
lon = -170
shrink, pad = 1, 0.02

# --- Configuration for each subplot ---
fields = [
    ("CT_e_VPR", cmo.matter_r, (1e-12, 1e-8),  "log", "Var. Production (°C^2/s)", "CT Meso. Variance Production", 200),
    ("SA_e_VPR", cmo.matter_r, (1e-13, 1e-10),  "log", "Var. Production (°C^2/s)", "SA Meso. Variance Production", 200),
    ("CT_m_VPR", cmo.matter_r, (1e-12, 1e-8),  "log", "Var. Production (°C^2/s)", "CT Micro. Variance Production", 100),
    ("SA_m_VPR", cmo.matter_r, (1e-13, 1e-10),  "log", "Var. Production (°C^2/s)", "SA Micro. Variance Production", 100),
    ("CT_meso_frac", cmo.balance, (0, 1),      "linear", "Mesoscale Fraction", "CT Meso. Fraction", 200),
    ("SA_meso_frac", cmo.balance, (0, 1),      "linear", "Mesoscale Fraction", "SA Meso. Fraction", 200),]

# --- Helper function to plot each panel ---
def add_hatching(ax, thickness):
    # Top region (0–thickness)
    ax.axhspan(0, thickness, facecolor='none', edgecolor='k', hatch='xxx', linewidth=0)

    # Bottom region (2000–2000-thickness)
    ax.axhspan(2000 - thickness, 2000, facecolor='none', edgecolor='k', hatch='xxx', linewidth=0)
    
def plot_panel(ax, varname, cmap, vlims, norm_type, cbar_label, title, is_bottom_row, thickness):
    da = ds_pbar_grid[varname].sel(LON=lon, method='nearest')

    # Choose the correct normalization
    if norm_type == "log":
        norm = colors.LogNorm(vmin=vlims[0], vmax=vlims[1])
    else:
        norm = colors.Normalize(vmin=vlims[0], vmax=vlims[1])

    da.plot(x='LAT', y='PRESSURE_mean', cmap=cmap, norm=norm, ax=ax, cbar_kwargs={'shrink': shrink, 'pad': pad, 'label': cbar_label})

    ax.set_xlim(-72, 58)
    ax.set_ylim(2000, 0)
    ax.set_ylabel("Pressure Mean (dbar)")

    # Handle x‑labels and tick labels
    if is_bottom_row:
        ax.set_xlabel("Latitude (°N)")
        ax.set_xticklabels(ax.get_xticklabels())
    else:
        ax.set_xlabel("")
        ax.set_xticklabels([])

    # Title inside the panel, in gray
    ax.text(0.5, 0.1, title, transform=ax.transAxes, ha='center', va='top', fontweight='bold', color='white', fontsize=plt.rcParams['axes.titlesize'])
    ax.set_title('')

    # Add hashed regions
    add_hatching(ax, thickness)

# --- Create figure and loop ---
fig, axes = plt.subplots(3, 2, figsize=(15, 11.25))
axes = axes.ravel()

for i, ax in enumerate(axes):
    var, cmap, vlims, norm_type, cbar_label, title, thickness = fields[i]
    is_bottom_row = i >=4
    plot_panel(ax, var, cmap, vlims, norm_type, cbar_label, title, is_bottom_row, thickness)

plt.subplots_adjust(wspace=0.3, hspace=0.1)
plt.savefig(f"quals_paper/supplemental/4_{ds_pbar_grid.sel(LON=lon, method='nearest').LON.data}.png")
print(f"lon: {ds_pbar_grid.sel(LON=lon, method='nearest').LON.data}")
```

### Results 7

```python
depth = 1800
shrink, pad = 0.9, 0.1

# --- Configuration for each subplot ---
fields = [
    ("F_heat_xy", cmo.deep_r, (1e1, 1e7), "log", "Heat Flux (W/m²)", "Eddy Heat Flux", "heat", 75),
    ("F_CT_xy",   cmo.deep_r, (1e-6, 10**(-2.5)), "log", "CT Flux (°C m/s)", "Eddy Temperature Flux", "CT", 75),
    ("F_SA_xy",   cmo.deep_r, (1e-7, 1e-3), "log", "SA Flux (g/kg/s)", "Eddy Salt Flux", "SA", 75),]


# --- Helper function to plot each panel ---
def plot_panel(ax, varname, cmap, vlims, norm_type, cbar_label, title, vec_type, scale):
    da = ds_pbar_grid[varname].sel(PRESSURE_mean=depth, method='nearest')

    # Choose normalization
    if norm_type == "log":
        norm = colors.LogNorm(vmin=vlims[0], vmax=vlims[1])
    else:
        norm = colors.Normalize(vmin=vlims[0], vmax=vlims[1])

    # --- Base map ---
    da.plot(x='LON', y='LAT', cmap=cmap, norm=norm, transform=ccrs.PlateCarree(), ax=ax, cbar_kwargs={'shrink': shrink, 'pad': pad, 'label': cbar_label})

    # --- Select vector components ---
    if vec_type == "CT":
        u = ds_pbar_grid['F_CT_x'].sel(PRESSURE_mean=depth, method='nearest')
        v = ds_pbar_grid['F_CT_y'].sel(PRESSURE_mean=depth, method='nearest')
    elif vec_type == "SA":
        u = ds_pbar_grid['F_SA_x'].sel(PRESSURE_mean=depth, method='nearest')
        v = ds_pbar_grid['F_SA_y'].sel(PRESSURE_mean=depth, method='nearest')
    elif vec_type == "heat":
        u = ds_pbar_grid['F_heat_x'].sel(PRESSURE_mean=depth, method='nearest')
        v = ds_pbar_grid['F_heat_y'].sel(PRESSURE_mean=depth, method='nearest')
    else:
        raise ValueError(f"Unknown vec_type: {vec_type}")

    # --- Meshgrid ---
    X = u['LON'].values
    Y = u['LAT'].values
    XX, YY = np.meshgrid(X, Y, indexing='ij')

    U = u.transpose('LON', 'LAT').values
    V = v.transpose('LON', 'LAT').values

    #scale to unit vectors
    mag = np.hypot(U, V)
    mag[mag == 0] = np.nan
    U_plot = U / mag
    V_plot = V / mag

    # --- Quivers ---
    q = ax.quiver(
        XX, YY, U_plot, V_plot, scale=scale,
        transform=ccrs.PlateCarree(),
        pivot='middle', width=0.0015,
        color='k'
    )

    # Add quiverkey
    #ax.quiverkey(q, X=0.75, Y=0.05, U=scale, label=f'{scale} gradient units', labelpos='E', color='k', labelcolor='k')

    # --- Decorations ---
    ax.coastlines()
    ax.gridlines(draw_labels=True, alpha=0.75)
    ax.set_title(title)

# --- Create figure and loop ---
fig, axes = plt.subplots(2, 2, figsize=(15, 7.5), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=-60)})
axes = axes.ravel()

for i, ax in enumerate(axes[:len(fields)]):
    var, cmap, vlims, norm_type, cbar_label, title, vec_type, scale = fields[i]
    plot_panel(ax, var, cmap, vlims, norm_type, cbar_label, title, vec_type, scale)

# Hide the extra subplot
for ax in axes[len(fields):]:
    ax.set_visible(False)

plt.subplots_adjust(wspace=0.15, hspace=0.1)
plt.savefig(f"quals_paper/supplemental/7_{ds_pbar_grid.sel(PRESSURE_mean=depth, method='nearest').PRESSURE_mean.data}.png")
```

### POSTER Figure 7

```python
depth = 500
shrink, pad = 0.9, 0.1

# --- Configuration for each subplot ---
fields = [
    ("F_heat_xy", cmo.deep_r, (1e1, 1e7), "log", "Heat Flux (W/m²)", "Eddy Heat Flux", "heat", 75),
    ("F_SA_xy",   cmo.deep_r, (1e-7, 1e-3), "log", "SA Flux (g/kg/s)", "Eddy Salt Flux", "SA", 75),]


# --- Helper function to plot each panel ---
def plot_panel(ax, varname, cmap, vlims, norm_type, cbar_label, title, vec_type, scale):
    da = ds_pbar_grid[varname].sel(PRESSURE_mean=depth, method='nearest')

    # Choose normalization
    if norm_type == "log":
        norm = colors.LogNorm(vmin=vlims[0], vmax=vlims[1])
    else:
        norm = colors.Normalize(vmin=vlims[0], vmax=vlims[1])

    # --- Base map ---
    da.plot(x='LON', y='LAT', cmap=cmap, norm=norm, transform=ccrs.PlateCarree(), ax=ax, cbar_kwargs={'shrink': shrink, 'pad': pad, 'label': cbar_label})

    # --- Select vector components ---
    if vec_type == "CT":
        u = ds_pbar_grid['F_CT_x'].sel(PRESSURE_mean=depth, method='nearest')
        v = ds_pbar_grid['F_CT_y'].sel(PRESSURE_mean=depth, method='nearest')
    elif vec_type == "SA":
        u = ds_pbar_grid['F_SA_x'].sel(PRESSURE_mean=depth, method='nearest')
        v = ds_pbar_grid['F_SA_y'].sel(PRESSURE_mean=depth, method='nearest')
    elif vec_type == "heat":
        u = ds_pbar_grid['F_heat_x'].sel(PRESSURE_mean=depth, method='nearest')
        v = ds_pbar_grid['F_heat_y'].sel(PRESSURE_mean=depth, method='nearest')
    else:
        raise ValueError(f"Unknown vec_type: {vec_type}")

    # --- Meshgrid ---
    X = u['LON'].values
    Y = u['LAT'].values
    XX, YY = np.meshgrid(X, Y, indexing='ij')

    U = u.transpose('LON', 'LAT').values
    V = v.transpose('LON', 'LAT').values

    #scale to unit vectors
    mag = np.hypot(U, V)
    mag[mag == 0] = np.nan
    U_plot = U / mag
    V_plot = V / mag

    # --- Quivers ---
    q = ax.quiver(
        XX, YY, U_plot, V_plot, scale=scale,
        transform=ccrs.PlateCarree(),
        pivot='middle', width=0.0015,
        color='k'
    )

    # Add quiverkey
    #ax.quiverkey(q, X=0.75, Y=0.05, U=scale, label=f'{scale} gradient units', labelpos='E', color='k', labelcolor='k')

    # --- Decorations ---
    ax.coastlines()
    ax.gridlines(draw_labels=True, alpha=0.75)
    ax.set_title(title)

# --- Create figure and loop ---
fig, axes = plt.subplots(1, 2, figsize=(15, 3.75), subplot_kw={'projection': ccrs.PlateCarree(central_longitude=-60)})
axes = axes.ravel()

for i, ax in enumerate(axes):
    var, cmap, vlims, norm_type, cbar_label, title, vec_type, scale = fields[i]
    plot_panel(ax, var, cmap, vlims, norm_type, cbar_label, title, vec_type, scale)

plt.subplots_adjust(wspace=0.15, hspace=0.1)
plt.savefig(f"quals_paper/results/POSTER_7_{ds_pbar_grid.sel(PRESSURE_mean=depth, method='nearest').PRESSURE_mean.data}.png")
```

### POSTER Figure 8

```python
lon = -28
shrink, pad = 1, 0.02

# --- Configuration for each subplot ---
fields = [
    ("F_heat_xy", cmo.deep_r, (1e3, 1e7), "log", "Heat Flux (W/m^2)", "2D Eddy Heat Flux", 200),
    ("F_SA_xy", cmo.deep_r, (1e-6, 1e-3), "log", "SA Flux (gm/kg/s)", "2D Eddy Salt Flux", 200),]

# --- Helper function to plot each panel ---
def add_hatching(ax, thickness):
    # Top region (0–thickness)
    ax.axhspan(0, thickness, facecolor='none', edgecolor='k', hatch='xxx', linewidth=0)

    # Bottom region (2000–2000-thickness)
    ax.axhspan(2000 - thickness, 2000, facecolor='none', edgecolor='k', hatch='xxx', linewidth=0)
    
def plot_panel(ax, varname, cmap, vlims, norm_type, cbar_label, title, is_bottom_row, thickness):
    da = ds_pbar_grid[varname].sel(LON=lon, method='nearest')

    # Choose the correct normalization
    if norm_type == "log":
        norm = colors.LogNorm(vmin=vlims[0], vmax=vlims[1])
    else:
        norm = colors.Normalize(vmin=vlims[0], vmax=vlims[1])

    da.plot(x='LAT', y='PRESSURE_mean', cmap=cmap, norm=norm, ax=ax, cbar_kwargs={'shrink': shrink, 'pad': pad, 'label': cbar_label})

    ax.set_xlim(-68, 62)
    ax.set_ylim(2000, 0)
    ax.set_ylabel("Pressure Mean (dbar)")

    # Handle x‑labels and tick labels
    if is_bottom_row:
        ax.set_xlabel("Latitude (°N)")
        ax.set_xticklabels(ax.get_xticklabels())
    else:
        ax.set_xlabel("")
        ax.set_xticklabels([])

    # Title inside the panel, in gray
    ax.text(0.5, 0.1, title, transform=ax.transAxes, ha='center', va='top', fontweight='bold', color='white', fontsize=plt.rcParams['axes.titlesize'])
    ax.set_title('')

    # Add hashed regions
    add_hatching(ax, thickness)

# --- Create figure and loop ---
fig, axes = plt.subplots(1,2, figsize=(15, 3.75))
axes = axes.ravel()

for i, ax in enumerate(axes):
    var, cmap, vlims, norm_type, cbar_label, title, thickness = fields[i]
    is_bottom_row = i >=0
    plot_panel(ax, var, cmap, vlims, norm_type, cbar_label, title, is_bottom_row, thickness)

plt.subplots_adjust(wspace=0.3, hspace=0.1)
plt.savefig(f"quals_paper/results/POSTER_8_{ds_pbar_grid.sel(LON=lon, method='nearest').LON.data}.png")
print(f"lon: {ds_pbar_grid.sel(LON=lon, method='nearest').LON.data}")
```
